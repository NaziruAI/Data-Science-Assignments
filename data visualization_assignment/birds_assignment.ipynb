# Basic data cleaning
### 

```python
bird_data.dropna(inplace=True)  # Drop rows with missing values
```

```python
# Build line chart
plt.figure(figsize=(10, 5))
plt.plot(bird_data['Unnamed: 0'], label='Unnamed: 0')
plt.title('Line Chart')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```

```python
# Build scatter plot
plt.figure(figsize=(10, 5))
plt.scatter(bird_data['Unnamed: 0'], bird_data['filepaths'], label='Scatter Plot', color='red')
plt.title('Scatter Plot')
plt.xlabel('Unnamed: 0')
plt.ylabel('Filepaths')
plt.legend()
plt.show()
```

```python
# Build bar chart
label_counts = bird_data['labels'].value_counts()
plt.figure(figsize=(10, 5))
label_counts.plot(kind='bar', color='green')
plt.title('Bar Chart - Label Counts')
plt.xlabel('Labels')
plt.ylabel('Count')
plt.show()
```
