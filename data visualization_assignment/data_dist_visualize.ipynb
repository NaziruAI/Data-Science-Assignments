# Data Distribution and Visualization

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
births_data = pd.read_csv('../data/births.csv')
```

# Explore the dataset

```python
print(births_data.head())
```

```python
print(births_data.info())
```

```python
# Basic statistics
print(births_data.describe())
```

```python
# Visualize the distribution of birth year using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(births_data['year'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Birth Year')
plt.xlabel('Birth Year')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize the distribution of birth month using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(births_data['month'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Birth month')
plt.xlabel('Birth Month')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize the distribution of birth days using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(births_data['day'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Birth Day')
plt.xlabel('Birth Day')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize the distribution of birth births using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(births_data['births'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Births')
plt.xlabel('Births')
plt.ylabel('Frequency')
plt.show()
```
